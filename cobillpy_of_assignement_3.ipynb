part 1

understanding the problem, you have two strings, we need to check if they are anagrams, so like we can spell one another with the same letters. So in this problem we'll need to rearrange the letter to see if they can match

input s and t are both lowercases letter.
out is either true or false
constraints are both strings are made up of lowercases letters, and the strings length are less than 50000

```python
# part 2

# I looked at the leetcode solution and used the if len(s) != len(t)

class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False

        sort_s = sorted(s)
        sort_t = sorted(t)

        if sort_s == sort_t:
            return True
        else:
            return False
```
